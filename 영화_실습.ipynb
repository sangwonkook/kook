#데이터

```python
import pandas as pd
import numpy as np
url1 = ('https://raw.githubusercontent.com/sd12832/MoViZ/master/' +
        'src/tmdb_5000_credits.csv')
df1 = pd.read_csv(url1)
print(df1.shape)
df1.head()
```

```python
url2 = ('https://raw.githubusercontent.com/sd12832/MoViZ/master/' +
        'src/tmdb_5000_movies.csv')
df2 = pd.read_csv(url2)
print(df2.shape)
df2.head()
```

```python
# prompt: df1의 컬럼명을 ['id', 'tittle', 'cast', 'crew']로 변경

df1.columns = ['id', 'tittle', 'cast', 'crew']
```

```python
# prompt: df2와 df1을 'id' 컬럼을 기준으로 병합 (merge)

df1.merge(df2, on='id')
```

```python
# prompt: 병합된 데이터프레임의 shape를 출력

merged_df = df1.merge(df2, on='id')
merged_df.shape
```

```python
# prompt: 컬럼명 전체를 보고싶어

merged_df.columns
```

#실습

```python
#평점 통계 계산
C = merged_df['vote_average'].mean()
m = merged_df['vote_count'].quantile(0.9)
C, m
```

```python
#투표수 상위 10%인 영화들만 분류
q_movies = merged_df.loc[merged_df['vote_count'] >= m]
q_movies.shape    # (481, 23)
```

```python
# prompt: 영화 추천 시스템을 위한 IMDB 가중평점 계산 코드를 작성해주세요.

def weighted_rating(x, m=m, C=C):
  v = x['vote_count']
  R = x['vote_average']
  return (v/(v+m) * R) + (m/(v+m) * C)

q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
q_movies = q_movies.sort_values('score', ascending=False)
q_movies[['title', 'vote_count', 'vote_average', 'score']].head(10)
```

```python
#평점 통계 계산
C = merged_df['vote_average'].mean()
m = merged_df['vote_count'].quantile(0.5)
C, m
```

```python
#평점 통계가 상위 50%이상이고 그중에 비영어권 언어 영화를 분류한것
f_movies = merged_df.loc[merged_df['vote_count'] >= m]
f_movies =f_movies.loc[merged_df['original_language'] != 'en']
f_movies.shape    # (298, 23)
```

```python
# prompt: 영화 추천 시스템을 위한 IMDB 가중평점 계산 코드를 작성해주세요.

# 외국어 영화 가중평점 계산
C = f_movies['vote_average'].mean()
m = f_movies['vote_count'].quantile(0.5)

f_movies['score'] = f_movies.apply(weighted_rating, axis=1)
f_movies = f_movies.sort_values('score', ascending=False)
f_movies[['title', 'vote_count', 'vote_average', 'score']].head(10)
```

#줄거리 기반 추천 알고리즘 생성

| 칼럼명                    | 설명                                                        |
| ---------------------- | --------------------------------------------------------- |
| `id`                   | 영화 고유 ID (숫자, 내부 식별자)                                     |

| `title`                | 영화 제목 (예: *Inception*)                                    |

| `tagline`              | 마케팅 슬로건 (예: "Your mind is the scene of the crime.")       |
| `genres`               | 장르 리스트 (예: Action, Drama 등)                               |
| `keywords`             | 핵심 키워드 (예: dream, subconscious, spy 등)                    |
| `cast`                 | 주요 배우 리스트 (예: Leonardo DiCaprio, Joseph Gordon-Levitt...) |
| `crew`                 | 감독, 작가, 프로듀서 등의 정보                                        |
| `budget`               | 제작비 (단위: USD, 보통 숫자)                                      |
| `revenue`              | 박스오피스 수익 (단위: USD)                                        |
| `runtime`              | 상영 시간 (단위: 분)                                             |
| `release_date`         | 개봉일 (형식: YYYY-MM-DD)                                      |
| `status`               | 제작 상태 (Released, Post Production 등)                       |
| `homepage`             | 영화 공식 홈페이지 URL                                            |
| `original_language`    | 원어 (예: en, fr, ko 등)                                      |
| `spoken_languages`     | 실제 대사에 쓰인 언어들                                             |
| `production_companies` | 제작사 리스트 (예: Warner Bros, Pixar 등)                         |
| `production_countries` | 제작국가 리스트 (예: 미국, 프랑스 등)                                   |
| `popularity`           | TMDB에서 정의한 인기 점수 (조회수, 활동 등 기반)                           |
| `vote_average`         | 평균 평점 (0\~10점 사이)                                         |
| `vote_count`           | 평점을 남긴 사람 수 (리뷰 수, 투표 수)                                  |


```python
# prompt: q_movies데이터에서 id, title, overview 칼럼만을 포함한 표를 만들어줘

q_movies[['id', 'title', 'overview']]
```

```python
# prompt: q_movies에서 'title'  칼럼 inception을 기준으로 전체 영화의 TF-IDF를 분석해줘

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# TF-IDF Vectorization
tfidf = TfidfVectorizer(stop_words='english')
q_movies['overview'] = q_movies['overview'].fillna('')  # Fill NaN with empty string
tfidf_matrix = tfidf.fit_transform(q_movies['overview'])

# Compute the cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# Construct a reverse map of indices and movie titles
indices = pd.Series(q_movies.index, index=q_movies['title']).drop_duplicates()

# Function that takes in movie title as input and outputs most similar movies
def get_recommendations(title, cosine_sim=cosine_sim):
  # Get the index of the movie that matches the title
  idx = indices[title]

  # Get the pairwise similarity scores of all movies with that movie
  sim_scores = list(enumerate(cosine_sim[idx]))

  # Sort the movies based on the similarity scores
  sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

  # Get the scores of the 10 most similar movies
  sim_scores = sim_scores[1:11]

  # Get the movie indices
  movie_indices = [i[0] for i in sim_scores]

  # Return the top 10 most similar movies
  return q_movies['title'].iloc[movie_indices]

# Get recommendations for '영화 제목'
print(get_recommendations('Fantastic Four'))
```

```python
# prompt: f_movies 데이터에서 'cast' , 'crew', 'keywords', 'genre' 데이터와 'title' 항목으로 구성된 표를 만들어줘

f_movies[['title', 'cast', 'crew', 'keywords', 'genres']]
```

```python
# prompt: # prompt: f_movies 데이터에서 'cast' , 'crew', 'keywords', 'genre' 데이터와 'title' 항목으로 구성 돼있고 score로 정렬된 표를 만들어줘

f_movies_selected = f_movies[['title', 'cast', 'crew', 'keywords', 'genres', 'score']].sort_values('score', ascending=False)
f_movies_selected.head(10)
```

```python
# prompt: # prompt: f_movies에서 'title'  칼럼 'Oldboy'를 기준으로 전체 영화의 'title', 'cast', 'crew', 'keywords', 'genres' Cos 유사성을 분석해줘

import ast

features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
  f_movies_selected[feature] = f_movies_selected[feature].apply(
      lambda x: ast.literal_eval(x) if isinstance(x, str) else []
  )


def get_director(x):
  for i in x:
    if i['job'] == 'Director':
      return i['name']
  return np.nan


def get_list(x):
  if isinstance(x, list):
    names = [i['name'] for i in x]
    if len(names) > 3:
      names = names[:3]
    return names
  return []


f_movies_selected['director'] = f_movies_selected['crew'].apply(get_director)

features = ['cast', 'keywords', 'genres']
for feature in features:
  f_movies_selected[feature] = f_movies_selected[feature].apply(get_list)


def clean_data(x):
  if isinstance(x, list):
    return [str.lower(i.replace(' ', '')) for i in x]
  else:
    if isinstance(x, str):
      return str.lower(x.replace(' ', ''))
    else:
      return ''


features = ['cast', 'keywords', 'genres', 'director']
for feature in features:
  f_movies_selected[feature] = f_movies_selected[feature].apply(clean_data)


def create_soup(x):
  return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])


f_movies_selected['soup'] = f_movies_selected.apply(create_soup, axis=1)

from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(f_movies_selected['soup'])

from sklearn.metrics.pairwise import cosine_similarity

cosine_sim2 = cosine_similarity(count_matrix, count_matrix)

f_movies_selected = f_movies_selected.reset_index()
indices2 = pd.Series(f_movies_selected.index, index=f_movies_selected['title'])


def get_recommendations2(title, cosine_sim=cosine_sim2):
  idx = indices2[title]
  sim_scores = list(enumerate(cosine_sim[idx]))
  sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
  sim_scores = sim_scores[1:11]
  movie_indices = [i[0] for i in sim_scores]
  return f_movies_selected['title'].iloc[movie_indices]


print(get_recommendations2('Oldboy'))
```
